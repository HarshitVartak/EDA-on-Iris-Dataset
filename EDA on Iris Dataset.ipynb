#### Scatter Plot

```python
sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species")
```

#### Joint Plot

```python
sns.jointplot(data=iris, x="SepalLengthCm", y="SepalWidthCm")
```

#### Box Plot

```python
sns.boxplot(data=iris, x="Species", y="PetalLengthCm")
```

#### Violin Plot

```python
sns.violinplot(data=iris, x="Species", y="PetalLengthCm")
```

#### Pair Plot

```python
sns.pairplot(data=iris, vars=["Id", "PetalLengthCm", "PetalWidthCm"], hue="Species")
```

#### Bar Plot

```python
sns.barplot(data=iris, x="Species", y="SepalWidthCm")
```

#### Swarm Plot

```python
sns.swarmplot(data=iris, x='PetalWidthCm',y='PetalLengthCm', hue = 'Species')
```

#### Dist plot

```python
sns.distplot(iris["SepalWidthCm"], bins=10, kde=True, color="red")
```
